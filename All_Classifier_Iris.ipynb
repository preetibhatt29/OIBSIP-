```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
import pandas as pd
```

```python
h = 0.02  # step size in the mesh

names = [
    "Nearest Neighbors",
    "Linear SVM",
    "RBF SVM",
    "Gaussian Process",
    "Decision Tree",
    "Random Forest",
    "Neural Net",
    "AdaBoost",
    "Naive Bayes",
    "QDA",
]

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
]
```

```python
X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [
    make_moons(noise=0.3, random_state=0),
    make_circles(noise=0.2, factor=0.5, random_state=1),
    linearly_separable,
]


```

```python
#Load Data

df = pd.read_csv('/content/drive/My Drive/OASIS/Iris.csv')
```

```python
df = df.drop(columns = ['Id'])
df = df.drop(columns = ['Species'])
df.head()
```

```python

X = df.drop(['Species1'], axis='columns')
Y = df.Species1

from sklearn.model_selection import train_test_split
#train, test = train_test_split(df, test_size=0.25, stratify=df['Species1'])

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, stratify=df['Species1'])
#x_train= train.drop(['Species1'], axis='columns')
#y_train = train.Species1
#x_test= test.drop(['Species1'], axis='columns')
#y_test = test.Species1
print("==================================================")
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
from pickleshare import print_function
# iterate over datasets
i=1
# iterate over classifiers
for name, clf in zip(names, classifiers):
    clf.fit(x_train, y_train)
    y_pred=clf.predict(x_test)
    score = clf.score(x_test, y_test)
    i=i+1
    print("%s = %.2f \t" %(name, score))
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred)) #shows all
    print("++++++++++++++++++++++++++++++++++++++++++++++++")

```
